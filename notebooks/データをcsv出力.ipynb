必要なライブラリをインポート

```python
import pandas as pd
import re
#文字列の決まった表現を検索するためのライブラリ。
```

ドライブ内にあるファイルを読み込む

```python
filename="message"
f = open(f'/content/drive/MyDrive/data/{filename}.txt', 'r')
datalist = f.readlines()
f.close()
#readlinesでテキストファイルを一行ずつ読み込む
datalist
```

```python
lis=[]
for data in datalist:
  m=re.match(r"202[0-3].[01][0-9].[0-3][0-9]",data)
  #保存したテキストデータはその日の先頭に日付が書かれていたので該当する日付を取得する。
  if m is not None:
    present_date=m.group()
    #m.group()で検索結果に合った文字列を取得
  a=data.split(" ")
  #文字列を空白で区切っているところだけ抽出する。
  if len(a)==3:
    a[0]=present_date+" "+a[0]
    lis.append(a)
    #a[0]には時間だけ入っているのでそれに日にちを追加する。

df=pd.DataFrame(lis,columns=["時間","名前","内容"])
df.to_csv(f"/content/drive/MyDrive/data/{filename}.csv")
```
