wikipediaの要素をデータフレームに変換

```python
import json
import pandas as pd
res = []
decoder = json.JSONDecoder()
for i in range(10,30):
  with open(f'wiki_{i}', 'r') as f:
    #fstringで文字列の中に変数iを代入している。
      line = f.readline()
      while line:
          res.append(decoder.raw_decode(line))
          line = f.readline()

lis=[]
for i, d in enumerate(res):
  lis.append(d[0]["text"])

df=pd.DataFrame(lis,columns=["content"])
#新たなデータフレームを作成
df
```

テキストデータを単語ごとに分解する

```python
import MeCab
import re
import pandas as pd
#MeCabのインスタンスを作成する
mecab = MeCab.Tagger('')

#最終的な結果を出力するためのData Frameを作成する
df1 = pd.DataFrame( columns=['word0','word1','word2','word3','word4','word5','word6'] )

#dfにセットしたcsvファイルを一行ずつ読み込む
for row, item in df.iterrows():
    #形態素解析結果をセットする変数。一行ずつ（opinionaireid単位）処理するため、処理前に一度クリアする
    result = ''
    #一行ずつ形態素解析を行い、単語ごとに分ける
    result = mecab.parse(item.content)
    #一行ごとに分けて変数にセットする
    lines = result.split("\n")
    #後ろから２列は不要のため削除する
    lines = lines[0:-2]
    #一行ずつに分けた形態素解析結果の変数を上から順に読み込む
    for words in lines:
        #タブとカンマで区切られているため、配列して新たな変数にセットする
        word = re.split('\t|,',words)
        #結果をData Frameにセットする
        df1 = df1.append({ 'word0':word[0], 'word1':word[1], 'word2':word[2], 'word3':word[3], 'word4':word[4], 'word5':word[5], 'word6':word[6]},ignore_index=True)


#全件表示
pd.set_option('display.max_rows', None)
#Data Frameの中身を表示する
df1
```

特徴量を抽出する関数(漢字の割合、敬語の割合、えもじの割合)

```python
def get_feature_value(df):
  lis=[]
  for index,content in zip(df.index,df["内容"]):
    str_amount=len(content)
    #文字列の長さ抽出
    emoji_parcent=emoji.emoji_count(content)/str_amount
    #文字列に対する絵文字の割合を出す。
    kanji_parcent=len(re.findall('[\u4E00-\u9FD0]', content))/str_amount
    #文字列に対する漢字の割合を出す。
    tagger=MeCab.Tagger()
    words=tagger.parse(content).split("\n")
    #一行ごとに形態素解析。splitをすることで各単語に分けることができる。
    words=words[:-2]
    #いらない要素があるので消す
    keigo_count=0
    for word in words:
      keigo_check=word.split("\t")[3]
      #3番目の要素に比較する敬語かどうか比較する要素が入っている
      if keigo_check in keigo_lis:
        #敬語がまとめてあるリストと要素を比較する。
        keigo_count+=1
      else:
        continue
    keigo_parcent=keigo_count/len(words)
    lis.append([kanji_parcent,keigo_parcent,emoji_parcent])
    df_kyousi=pd.DataFrame(lis,columns=["漢字の割合","敬語の割合","絵文字の割合"])
  df=pd.concat([df, df_kyousi], axis=1)
  return df   


      

df_kyousi=get_feature_value(df_kyousi)

df_kyousi
```
